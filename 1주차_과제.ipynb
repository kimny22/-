#다양한 제품들과 서비스 속 인공지능 기술
###- 김나영
 일상 속에서 인공지능 기술이 적용된 분야가 다양하게 있습니다. 다음은 언어, 음성, 이미지, 자율주행, 이 4가지 분야에서 제가 흥미로웠던 제품과 서비스입니다.

**1. 언어 분야 - 파파고**

파파고는 네이버에서 제공하는 AI 번역 서비스입니다. 네이버에서 자체 개발한 인공신경망을 기반으로 하였고 현재 한국어, 영어, 일본어, 중국어, 프랑스어 등 다양한 언어를 번역할 수 있습니다. 이제 점점 많은 번역기들이 발전하여 예전보다 자연스럽게 번역을 해주는데, 특히 파파고가 다른 번역 서비스보다 흥미로웠던 점은 트렌드를 따라간다는 점입니다.
 
기본적으로 어색하지 않게 번역을 잘 해주기도 하지만, 신조어도 해석한다는 점이 다른 번역기와 차별화가 됩니다. 예를 들어, '아재개그', '띵작'같이 다소 외국어로 번역하기 애매한 단어들도 'dad joke', 'rnasterpiece'로 적절히 번역하는 것을 볼 수 있습니다. 파파고가 신조어나 유행어를 잘 번역하는 이유는 사람들이 네이버 서비스를 이용하면서 쓰는 단어를 기반으로 데이터를 습득하기 때문입니다. 네이버의 동영상 서비스인 'V LIVE'에서 팬들이 쓴 번역 자막이 파파고에게 학습된다고 합니다.

파파고가 배우는 데이터들이 대부분 외국이 아닌 한국에서 만들어진 데이터라서 아무래도 한국어를 가장 잘 번역하는 번역기가 아닐까 싶습니다.

**2. 음성 분야 - 빅스비**

빅스비는 삼성전자의 AI 음성 비서로, 갤럭시S8 이후의 스마트폰에 모두 탑재되어 있습니다. 딥러닝 기술을 바탕으로 개발되어 사용자들의 데이터가 많이 쌓인 지금, 초기보다 음성인식이 더 잘 되고 있는 것을 볼 수 있습니다. 실제로 저도 빅스비를 유용하게 사용하고 있는데, 가장 큰 이유는 정확한 음성인식과 문장이해력입니다.

아침에 알람이 울리고 있을 때 잠이 덜 깬 작은 목소리로 빅스비에게 "좀 이따 깨워줘."라고 하면 5분 뒤로 알람 설정합니다. 이를 통해, 빅스비가 다소 알아듣기 힘들 수 있을 음성도 정확하게 인식하고 사용자가 원하는 것을 잘 캐치하는 이해력도 가졌음을 알 수 있습니다. 또한 다른 AI 음성 비서와 달리, 문장의 끝을 '-해', '-해줘', '-해줄 수 있을까', '-해주라' 등등 어떤 식으로 말하는지와 상관없이 모든 말을 이해하며, 여러 문장이 합쳐진 복잡한 명령도 잘 수행합니다.

지금의 빅스비 성능도 저는 어느 정도 만족하는데, 앞으로 데이터가 더 많이 쌓이면 얼마나 더 발전할지 기대가 됩니다.

**3. 이미지 분야 - 딥드림생성기**

딥드림생성기는 구글에서 개발한 프로그램으로, 웹상에서 생성기를 사용해 추상적인 그림을 그리는 AI입니다. 사용자가 원하는 사진을 올리고 스타일을 선택하면 그 스타일대로 재해석하여 사진을 왜곡시키는 프로그램입니다. 예를 들어, 풍경화 사진을 올리고 스타일로는 고흐의 그림을 선택하면 기존의 사진이 고흐 풍으로 재해석되어 나옵니다.

2016년에는 딥드림생성기로 그린 작품 여러 개가 경매에서 비싼 값에 팔리기도 하였습니다. 그만큼 사람이 직접 그린 것 같은 수준으로 추상적인 그림을 잘 그렸다고 볼 수 있습니다. 예술 분야는 인간의 영역이라고만 여겼는데, 점점 그 벽이 허물어지는 걸 보면서 과학의 발전을 체감할 수 있습니다.

**4. 자율주행 분야 - 록소**

록소는 페덱스의 라스트 마일용 자율 배달 로봇입니다. 계단을 오를 수 있어서 고객의 집 문 바로 앞까지 택배를 가져다줄 수 있으며, 현재 시제품 개발이 끝난 상태지만 그림자가 있을 때 오작동 가능성이 있어 빠르게 수정될 예정입니다.

이 로봇이 등장한 배경에는 현재 전세계적으로 유행하는 코로나19가 있습니다. 코로나19로 인해 사회적 거리두기를 하면서 식료품 산업계에 인력이 부족해졌고 그로 인해 시급한 과제로 떠오른 것입니다. 사실 록소 외에도 마트 바닥을 청소하는 로봇, 물류창고에서 상품들을 옮기는 로봇 등의 자율주행 로봇이 최근 도입되고 있습니다.

코로나19 이후 인공지능 로봇의 필요성이 부각되고 있는 것 같습니다. 몇 년은 더 걸릴 것이라고 예상했던 로봇이 지금 당장 쓰이게 됐다는 것이 흥미로웠습니다.
